```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/borjagq/Desktop/results.csv')

data = df.to_numpy()
```

```python
df.head()
```

```python
df.columns = ['epoch', 'train_box_loss', 'train_obj_loss', 'train_cls_loss', 'metrics_precision', 'metrics_recall', 'metrics_mAP_0.5', 'metrics_mAP_0.5:0.95', 'val_box_loss', 'val_obj_loss', 'val_cls_loss', 'x_lr0', 'x_lr1', 'x_lr2']
```

```python
train_box_loss = df['train_box_loss']
train_obj_loss = df['train_obj_loss']
train_cls_loss = df['train_cls_loss']
val_box_loss = df['val_box_loss']
val_obj_loss = df['val_obj_loss']
val_cls_loss = df['val_cls_loss']
```

```python
plt.plot(train_box_loss.to_numpy(), label='train_box_loss')
plt.plot(train_obj_loss.to_numpy(), label='train_obj_loss')
plt.plot(train_cls_loss.to_numpy(), label='train_cls_loss')
plt.plot(val_box_loss.to_numpy(), label='val_box_loss')
plt.plot(val_obj_loss.to_numpy(), label='val_obj_loss')
plt.plot(val_cls_loss.to_numpy(), label='val_cls_loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.legend(loc='upper right')

plt.show()
```

```python
metrics_precision = df['metrics_precision']
metrics_recall = df['metrics_recall']
```

```python
plt.plot(metrics_precision.to_numpy(), label='Precision')
plt.plot(metrics_recall.to_numpy(), label='Recall')

plt.xlabel('Epochs')
plt.ylabel('Score (%)')

plt.legend(loc='upper right')

plt.show()
```
